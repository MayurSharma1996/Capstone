Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
df1 = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m2_survey_data.csv")
df1.columns
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
df= pd.read_sql_query(QUERY,conn)
df.head(5)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt

distribution_curve= df1['ConvertedComp']
distribution_curve= distribution_curve.dropna()

plt.figure(figsize=(8, 6))
plt.hist(distribution_curve, bins=20, color='yellow')

plt.title('Salary Distribution')
plt.xlabel('ConvertedComp (Salary)')
plt.ylabel('Frequency')

plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
import seaborn as sns

distribution_age= df1['Age']
distribution_age= distribution_age.dropna()

plt.figure (figsize= (8,6))
sns.boxplot (data= distribution_age, color= 'yellow')
plt.title ('Age Distribution Table')
plt.xlabel ('Age')
plt.ylabel ('Frequency')

plt.grid (True)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

distribution_age= df1['Age']
distribution_age= distribution_age.dropna()

workweekhours= df1['WorkWeekHrs']
workweekhours= workweekhours.dropna()

plt.figure (figsize= (8,6))
sns.scatterplot (x= distribution_age, y=workweekhours, color= 'blue')
plt.title ('Age Distribution to Work Week Hours')

plt.grid (True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
x = df1['WorkWeekHrs']
y = df1['CodeRevHrs']
bubble_size = df1['Age']

plt.figure(figsize=(10, 8))

plt.scatter(x, y, s=bubble_size, alpha=0.5, c='blue', cmap='viridis')

plt.title('Bubble Plot of WorkWeekHrs vs. CodeRevHrs')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.colorbar(label='Age')

plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
database_desire_next_year = df1['DatabaseDesireNextYear'].str.split(';', expand=True).stack()

top_5_databases = database_desire_next_year.value_counts().head(5)

plt.figure(figsize=(8, 8))
plt.pie(top_5_databases, labels=top_5_databases.index, autopct='%1.1f%%', startangle=140)

plt.title('Top 5 Databases Respondents Wish to Learn Next Year')

plt.axis('equal')

plt.show()

```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
filtered_df = df1[(df1['Age'] >= 30) & (df1['Age'] <= 35)]

median_data = filtered_df.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()

plt.figure(figsize=(10, 6))

plt.bar(median_data.index, median_data['WorkWeekHrs'], label='WorkWeekHrs')

plt.bar(median_data.index, median_data['CodeRevHrs'], bottom=median_data['WorkWeekHrs'], label='CodeRevHrs')

plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')

plt.legend()

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
filtered_df = df1[(df1['Age'] >= 45) & (df1['Age'] <= 60)]

median_data = filtered_df.groupby('Age')[['ConvertedComp']].median()

plt.figure(figsize=(10, 6))
plt.plot(median_data.index, median_data.values, marker='o', linestyle='-')

plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.title('Median ConvertedComp for Ages 45-60')

plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

main_branch= df1['MainBranch']

plt.figure(figsize=(10, 6))
plt.barh(main_branch.index, main_branch.values, linewidth=10)

plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.title('Distribution of MainBranch')

plt.grid(axis='x')
plt.show()
```
